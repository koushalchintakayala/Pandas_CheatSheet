```python
import numpy as np
import pandas as pd
```

# **1. Creating a DataFrame**

**From Dictionary**

```python
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}
df = pd.DataFrame(data)
```

```python
df
```

**From List of Lists**

```python
data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]
df2 = pd.DataFrame(data_list)
```

```python
df2
```

```python
col = ["Name","Age","City","Salary"]
df2 = pd.DataFrame(data_list, columns =col)
```

```python
df2
```

# **2. Selection and Indexing of Columns**

**Single Column**

```python
df2["Name"]
```

**Multiple Columns**

```python
df2[["Name", "City"]]
```

# **3. Creating a new Column**

```python
df2["Designation"] = ["Doctor", "Eng.", "Doctor", "Eng."]
df2
```

# **4. Removing Columns**

**Drop Row**

```python
df2.drop(0, axis=0)   # removes row index 0
```

**Drop Column**

```python
df2.drop("Salary", axis=1)
```

```python
df2
```

```python
df2.drop("Salary", axis=1, inplace=True)  #use inplace=True, to modify original DataFrame
```

```python
df2
```

# **5. Selecting Rows**

**Using loc (label-based)**

```python
df2.loc[[0, 1]]
```

**Using iloc (index-based)**

```python
df2.iloc[3]
```

# **6. Selecting Rows + Columns Together**

```python
df2.loc[[0, 1], ["City", "Designation"]]  #Single-step selection (rows + columns together) / Faster + safer (direct access, no intermediate copy)
```

```python
df.loc[[0,1]][["City","Salary"]]   #Two-step selection (first rows, then columns) / Slightly slower + risky (creates temporary DataFrame, can cause issues in some cases)
```

# **7. Conditional Selection (Filtering)**

```python
df2
```

```python
#I only want to see those people whose age is above 30
```

```python
df2[df2["Age"] > 30]  #Single Condition
```

```python
#I only want poeple whose age is above 30 and their city must be paris
```

```python
df2[(df2["Age"] > 30) & (df2["City"] == "Paris")]  #Multiple Conditions
```
