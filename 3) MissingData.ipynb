```python
import numpy as np
import pandas as pd
```

# **1. Finding Missing Data**

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna()        # or df.isnull()   It Returns DataFrame of True/False
```

```python
df.isna().sum()      # Count missing values per column
```

```python
df.isna().any()      # Check if any missing values exist
```

```python
df.isna().sum().sum()    # Total missing values in entire dataset
```

# **2. Removing Missing Data**

```python
df
```

**Drop Rows**

```python
df.dropna()    # Removes rows with any NaN
```

**Drop Columns**

```python
df.dropna(axis=1)  # Removes columns with any NaN
```

**Threshold Concept**

```python
df.dropna(thresh=3)      # Keep rows with at least 3 non-NaN values
```

**Drop Based on Specific Columns**

```python
df.dropna(subset=['A', 'C'])   # Drops rows where A or C is NaN
```

# **3. Filling the missing Data**

```python
df.fillna(0)    # Fill with Constant
```

```python
df.fillna({'A':0, 'B':100, 'C':300})     # Fill with Different Values
```

```python
df.fillna(df.mean())  # Fill with Mean
```

```python
df.fillna(df.median())    # Fill with median
```

```python
df.fillna(df.mode().iloc[0])      # Fill with Mode
```

Interview Tip:

*   Mean → Normal distribution
*   Median → Skewed data
*   Mode → Categorical data

# **4. Forward / Backward Fill**

```python
df.fillna(method='ffill')   # Forward Fill (propagate previous value)
```

```python
df.fillna(method='bfill')     # Backward Fill
```

```python
df.ffill()
```

```python
df.bfill()
```

# **5. Interpolation**

```python
df.interpolate()    # Fills missing values using trend (linear by default)
```

# **6. Important Edge Cases**

```python
df['A'].isna().sum()     # Check Missing in Specific Column
```

```python
df.replace(0, np.nan)     # Replace Specific Values with NaN
```

```python
df.fillna(0, inplace=True)     # Inplace Modification
```

# **7. When to Drop vs Fill (Concept Question)**

| Situation                | What to Do     |
| ------------------------ | -------------- |
| Few missing values       | Drop           |
| Many missing values      | Fill           |
| Important column         | Fill carefully |
| Entire column mostly NaN | Drop column    |
