```python
import numpy as np
import pandas as pd
```

# **1. Merging 2 dataframes**

```python
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
employees
```

```python
salaries
```

**1. Inner Join**

```python
pd.merge(employees,salaries,on='employee_id',how ='inner')  # Common rows only.
```

**2. Left Join**

```python
pd.merge(employees,salaries,on='employee_id',how='left')
```

**3. Right Join**

```python
pd.merge(employees,salaries,on='employee_id',how='right')
```

**4. Outer Join**

```python
pd.merge(employees,salaries,on='employee_id',how='outer')
```

**5. Merge with Indicator**

```python
pd.merge(
  employees,
  salaries,
  on='employee_id',
  how='outer',
  indicator=True             # Shows source of rows.
)
```

**6. Selecting Custom Suffixes**

```python
salaries['department']=['HR','IT','Finance','Admin','Sales']
```

```python
pd.merge(
 employees,
 salaries,
 on='employee_id',
 suffixes=('_emp','_sal')
)
```

**7. Merge on Multiple Columns**

```python
pd.merge(
 employees,
 salaries,
 on=['employee_id'],
 how='inner'
)
```

**8. Cross Join**

```python
pd.merge(
 employees,
 salaries,              # Cartesian product.
 how='cross'            # Every employee with every salary row.
)
```

**9. Validate Merge**

```python
pd.merge(
 employees,
 salaries,
 on='employee_id',
 validate='one_to_one'        # Both tables must have unique values in the merge key.
)
```

```python
pd.merge(
 employees,
 salaries,
 on='employee_id',
 validate='one_to_many'      # Left table key unique, right table may have duplicates.
)
```

```python
pd.merge(
 employees,
 salaries,
 on='employee_id',
 validate='many_to_one'      # Left table can have duplicates, right table unique.
)
```

```python
pd.merge(
 employees,
 salaries,
 on='employee_id',
 validate='many_to_many'      # Duplicates allowed in both tables.
)
```

**Real-world mapping for validate merge**



*   Employee ↔ Salary → one_to_one
*   Customer ↔ Orders → one_to_many
*   Transactions ↔ Product Master → many_to_one
*   Students ↔ Courses (many enrollments) → many_to_many

Validate helps detect wrong merges and duplicate key issues







# **2. Concatenation of 2 dataframes**

```python
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2']
})

df2 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5'],
    'C': ['C3', 'C4', 'C5']
})
```

```python
df1
```

```python
df2
```

**Row Wise**

```python
pd.concat([df1,df2])
```

**Reset Index**

```python
pd.concat(
 [df1,df2],
 ignore_index=True
)
```

**Column Wise**

```python
pd.concat(
 [df1,df2],
 axis=1           # Column Wise
)
```

**Keys (Multi Index)**

```python
pd.concat(
 [df1,df2],
 keys=['first','second']
)
```

# **3. Joining data frames**
Index based.

```python
df1 = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie']
}, index=[1, 2, 3])

df2 = pd.DataFrame({
    'score': [85, 90, 75]
}, index=[2, 3, 4])
```

```python
df1
```

```python
df2
```

**Default Left Join**

```python
df1.join(df2)
```

**Outer Join**

```python
df1.join(df2,how='outer')
```

**Right Style Using Join**

```python
df2.join(df1)
```

**Join with Suffix**

```python
df2['name']=['X','Y','Z']     # Suppose both have "name"
```

```python
df1.join(
df2,
lsuffix='_left',
rsuffix='_right'
)
```

# **4) Handle Missing Data After Merge**

```python
pd.merge(
 employees,
 salaries,
 how='outer'
).fillna(0)    # replaces missing values with 0, keep all rows
```

```python
pd.merge(
 employees,
 salaries,
 how='outer'
).dropna()      # remove rows having missing values
```

# **5) Remove Duplicate Rows After Combining**

```python
pd.concat([df1,df2]).drop_duplicates()
```

# **6) Merge on Different Column Names**

```python
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

# DataFrame 2: Salary information
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
salaries
```

```python
salaries.columns = ['emp_id','salary','bonus']
```

```python
salaries
```

```python
pd.merge(
 employees,
 salaries,
 left_on='employee_id',
 right_on='emp_id'
)
```
