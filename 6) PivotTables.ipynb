````python
import numpy as np
import pandas as pd
````

# **Pivot Tables**

````python
data = {
    'Date': pd.date_range('2023-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] * 5,
    'Region': ['East', 'West', 'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West',
               'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Sales': np.random.randint(100, 1000, 20),
    'Units': np.random.randint(10, 100, 20),
    'Rep': ['John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary',
            'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice']
}

df = pd.DataFrame(data)

df['Month'] = df['Date'].dt.month_name()
df['Quarter'] = 'Q' + df['Date'].dt.quarter.astype(str)
df
````

# **1. Basic Pivot Table**

````python
pd.pivot_table(df, values="Sales", index="Region", columns="Product", aggfunc="median")
````

# **2. Multiple Aggregations**

````python
pd.pivot_table(df,
               values="Sales",
               index="Region",
               columns="Product",                      # It shows you know multi-metric analysis
               aggfunc=['mean', 'sum', 'max'])         # Multi-level column headers (mean, sum, max)
````

# **3. Multiple Values**

````python
pd.pivot_table(df,
               values=['Sales', 'Units'],                # It produces Hierarchical columns (Sales, Units)
               index='Region',
               columns='Product')
````

# **4. Fill Missing Values**

````python
pd.pivot_table(df,
               values="Sales",
               index="Region",
               columns="Product",
               fill_value=0)                       # It avoids NaN in reports
````

# **5. Margins (Grand Total)**

````python
pd.pivot_table(df,
               values="Sales",
               index="Region",
               columns="Product",
               aggfunc="sum",
               margins=True)                    # It adds all row & column (totals)
````

# **6. Group by Multiple Columns**

````python
pd.pivot_table(df,
               values="Sales",
               index=['Region', 'Rep'],         # Use case: Region + Representative analysis
               columns='Product',
               aggfunc='sum')
````

# **7. Using Month / Quarter**

````python
pd.pivot_table(df,
               values="Sales",
               index='Month',           # Time-based analysis = real-world scenario
               columns='Product',
               aggfunc='sum')
````

````python
pd.pivot_table(df,
               values="Sales",
               index='Quarter',         # Time-based analysis = real-world scenario
               columns='Region',
               aggfunc='mean')
````

# **8. Custom Aggregation**

````python
pd.pivot_table(df,
               values="Sales",
               index="Region",
               columns="Product",
               aggfunc=lambda x: x.max() - x.min())     # It gives range of sales per group
````

# **9. Pivot vs Pivot Table**

**Pivot (No aggregation)**



```
df.pivot(index="Region", columns="Product", values="Sales")
```

It fails if duplicate entries exist


**Pivot Table (Aggregation allowed)**

````python
pd.pivot_table(df, values="Sales", index="Region", columns="Product")    # Handles duplicates using aggregation
````

# **CROSSTAB (Frequency Analysis)**

# **10. Basic Crosstab**

````python
pd.crosstab(df['Region'], df['Product'])    # Counts occurrences
````

# **11. Normalization**

````python
pd.crosstab(df['Region'], df['Product'], normalize='index')       # Shows Row-wise percentage
````

````python
pd.crosstab(df['Region'], df['Product'], normalize='columns')       # Shows Column-wise percentage
````

````python
pd.crosstab(df['Region'], df['Product'], normalize=True)          # Shows Overall percentage
````

# **12. Crosstab with Aggregation**

````python
pd.crosstab(df['Region'],
            df['Product'],
            values=df['Sales'],
            aggfunc='sum')              # Now behaves like pivot table
````

# **13. Add Margins in Crosstab**

````python
pd.crosstab(df['Region'],
            df['Product'],
            margins=True)
````

You should be confident in:

1. Difference between:
pivot() vs pivot_table()

2. Handling:
Missing values (fill_value) vs
Totals (margins)

3. Multi-level:Multiple values vs
Multiple aggfunc

4. Time-based:
Month, Quarter

5. Crosstab:
Frequency vs percentage &
Aggregation inside crosstab
